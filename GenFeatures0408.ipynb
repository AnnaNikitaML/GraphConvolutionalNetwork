```python
def mean_angle(row):
    if row.num_bonds <= 1:
        return 0.
    
    angles = []
    mol_name = row.name[0]
    for i in range(row.num_bonds):
        idx1 = row.bond_to[i]
        row1 = strucs.loc[mol_name, idx1]
        x1,y1,z1 = row1.x - row.x, row1.y - row.y, row1.z - row.z
        r1 = np.sqrt(x1**2 + y1**2 + z1**2)
        for j in range(i + 1, row.num_bonds):
            idx2 = row.bond_to[j]
            row2 = strucs.loc[mol_name, idx2]
            x2,y2,z2 = row2.x - row.x, row2.y - row.y, row2.z - row.z
            r2 = np.sqrt(x2**2 + y2**2 + z2**2)
            
            angles.append(np.arccos( (x1*x2 + y1*y2 + z1*z2)/(r1*r2)))
    return np.mean(angles)*180.0/np.pi
            
    
    
```

```python
from scipy import special
def multiplicity(bond, dist):
    if ('H' in bond) or ('F' in bond):
        return 1.
    
    if bond == 'CC':
        if dist < 1.285:
            return 3.
        else:
            return 2. - np.clip(special.expit( (dist - 1.44) * 50) * 1.02 - 0.01, 0, 1)
        
    if (bond == 'CN') or (bond == 'NC'):
        if dist < 1.2:
            return 3.
        else:
            return 2. - np.clip(special.expit( (dist - 1.335) * 100) * 1.02 - 0.01, 0, 1)
        
    if (bond == 'CO') or (bond == 'OC'):
        if dist < 1.285:
            return 2.
        else:
            return 1.
    
    if bond == 'NN':
        return 2 - np.clip(special.expit( (dist - 1.28 ) * 80) * 1.02 - 0.01, 0, 1)
    
    if (bond == 'NO') or (bond == 'ON'):
        if dist > 1.30:
            return 1.
        else:
            return 2.5 - np.clip(special.expit( (dist - 1.24 ) * 100) * 1.02 - 0.01, 0, 1)
            
        
    
            
        
```

```python
def matrix_multiplicity(A, Type, Dist):
    M = np.zeros_like(A, dtype=np.float32)
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            if (A[i,j]):
                M[i,j] = multiplicity(Type[i,j], Dist[i,j])
                
    return M
    
```
